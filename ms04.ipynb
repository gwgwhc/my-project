```python
from iqtools import *
import numpy as np
%matplotlib inline
```

```python
#importing data
filename = 'tiqfiles/fumikara.tiq'
iqdata = get_iq_object(filename)
print(f"""nsamples: {iqdata.nsamples_total} 
sample frequency: {iqdata.fs}""")
```

```python
nsamples = iqdata.nsamples_total
lframes = 1024 # |number of bins
nframes = int(nsamples/lframes) #setting nframes to match number of bins over data
iqdata.read(nframes,lframes)
```

```python
iqdata.window='hamming'
iqdata.method='mtm'
xx, yy, zz = iqdata.get_spectrogram(nframes=nframes, lframes=lframes)
plot_spectrogram(xx,yy,zz,title='')
plt.savefig(f'ms04_{str(lframes)}.pdf')
```

```python
ff,pp,_=iqdata.get_fft()
plot_spectrum(ff,pp)
```
